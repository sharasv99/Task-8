```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from deep_translator import GoogleTranslator
```

```python
data = pd.read_csv('English.csv')
data.head()
```

```python
data["English words/sentences"] = data["English words/sentences"].str.replace("[^a-zA-Z0-9]", " ") 
data["English words/sentences"]
```

```python
data.head(50)
```

```python
def Translate_Es(rev):
    translation_tokenized = GoogleTranslator(source='english', target='es').translate(rev)
    return translation_tokenized
```

```python
Trans_Spanish_Words = [Translate_Es(r) for r in data["English words/sentences"].head(50)]
Trans_Spanish_Words
```

```python
Trans_Spanish_Sentences = [Translate_Es(r) for r in data["English words/sentences"].tail(50)]
Trans_Spanish_Sentences
```

```python
def Translate(rev):
    translation_tokenized = GoogleTranslator(source='english', target='fr').translate(rev)
    return translation_tokenized
```

```python
Trans_French_Words = [Translate(r) for r in data["English words/sentences"].head(50)]
Trans_French_Words
```

```python
Trans_French_Sentence = [Translate(r) for r in data["English words/sentences"].tail(50)]
Trans_French_Sentence
```

```python
def Translate_De(rev):
    translation_tokenized = GoogleTranslator(source='english', target='de').translate(rev)
    return translation_tokenized
```

```python
Trans_German_Words = [Translate_De(r) for r in data["English words/sentences"].head(50)]
Trans_German_Words
```

```python
Trans_German_Sentence = [Translate_De(r) for r in data["English words/sentences"].tail(50)]
Trans_German_Sentence
```
